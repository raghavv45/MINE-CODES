Transform Data:
Standardize formats

```python
data1['BookingEndDateTime (Month / Day / Year)'] = pd.to_datetime(
    data1['BookingEndDateTime (Month / Day / Year)'],
    format='%d-%b-%y'
)
```

Transform Data:
Normalize or scale numerical features if needed (StandardScaler from sklearn).


```python
#Imporitng Important library
from sklearn.preprocessing import StandardScaler
```

```python
# Identify numerical columns
numerical_columns = data1.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the numerical features
data1[numerical_columns] = scaler.fit_transform(data1[numerical_columns])
```
