```python
import pandas as pd

df=pd.read_excel("/content/Classes June 2018 Cleaned and preprocessed data.xlsx")
df1=pd.read_excel("/content/Cleaned and Preprocessed data Classes april-May 2018.xlsx")
```

```python
import datetime

def calculate_price(booking_date):
  """
  Calculates the price based on the booking date.

  Args:
    booking_date: A datetime object representing the booking date.

  Returns:
    The price as a float.
  """

  # Define the pricing rules
  weekday_price = 20
  weekend_price = 30
  peak_season_price = 40

  # Check if the booking date is during peak season
  # Convert peak season dates to pandas Timestamp objects for comparison
  peak_season_start = pd.Timestamp(datetime.date(2018, 7, 1))
  peak_season_end = pd.Timestamp(datetime.date(2018, 8, 31))
  if peak_season_start <= booking_date <= peak_season_end:
    return peak_season_price

  # Check if the booking date is a weekend
  # Extract the date from the Timestamp object for weekday comparison
  if booking_date.date().weekday() in [5, 6]:
    return weekend_price

  # Otherwise, it's a weekday
  return weekday_price

# Apply the function to the BookingEndDateTime column
df['Price'] = df['BookingEndDateTime (Month / Day / Year)'].apply(calculate_price)
df1['Price'] = df1['BookingEndDateTime (Month / Day / Year)'].apply(calculate_price)

# Print the dataframes
print(df)
print(df1)
```

```python
# Create a function to simulate booking dates
def generate_booking_dates(start_date, end_date, num_bookings):
  """
  Generates a list of random booking dates within a given date range.

  Args:
    start_date: A datetime object representing the start of the date range.
    end_date: A datetime object representing the end of the date range.
    num_bookings: The number of booking dates to generate.

  Returns:
    A list of datetime objects representing the booking dates.
  """

  import random
  dates = []
  for _ in range(num_bookings):
    random_date = start_date + (end_date - start_date) * random.random()
    dates.append(random_date)
  return dates

# Generate booking dates for June 2018
booking_dates_june = generate_booking_dates(
  start_date=pd.Timestamp(datetime.date(2018, 6, 1)),
  end_date=pd.Timestamp(datetime.date(2018, 6, 30)),
  num_bookings=100
)

# Generate booking dates for April-May 2018
booking_dates_april_may = generate_booking_dates(
  start_date=pd.Timestamp(datetime.date(2018, 4, 1)),
  end_date=pd.Timestamp(datetime.date(2018, 5, 31)),
  num_bookings=100
)

# Apply the pricing function to the generated booking dates
prices_june = [calculate_price(date) for date in booking_dates_june]
prices_april_may = [calculate_price(date) for date in booking_dates_april_may]

# Calculate the average price for each month
average_price_june = sum(prices_june) / len(prices_june)
average_price_april_may = sum(prices_april_may) / len(prices_april_may)

# Print the results
print(f"Average price for June: {average_price_june}")
print(f"Average price for April-May: {average_price_april_may}")

# Compare the average prices to evaluate the effectiveness of the pricing adjustments
# If the average price for June is higher than for April-May, the pricing adjustments were effective
```
